```python
import xtrack as xt
```

#### Load lattice

```python
env = xt.Environment()
env.call('../fccee_z_lattice.py')
```

#### Define a line from the element

```python
env.new_line(name='ccs_xl',  components=[
    2 * ['scrabl'], 'lx0', 'qd20l', 'd8l', 'qf19l', 'd8l', 'qd18l', 'd8l', 'qf17l', 'd8l',
    'qd16l', 'lx0', 'b7l', 'lx0', 'qf15l', 'lx0', 'b7l', 'lx0', 'qd14l', 'd7l', 'qf13l',
    'lx0', 'decfl', 4 * ['sfx2l'], 'lx0', 'qx0l', 'd7l', 'qx1l', 'lx0', 'b6l', 'lx0',
    'qx2l', 'ipimag4', 'qx2l', 'lx0', 'b6l', 'lx0', 'qx1l', 'd7l', 'qx0l', 'lx0',
    4 * ['sfx1l'], 'decfl', 'lx0', 'qf12l', 'd7l', 'qd11l', 'lx0', 'b5l', 'lx0', 'oct2l',
    2 * ['sfm2l'], 'lx0'])
```

```python
# Inspect using survey
env['ccs_yl'].survey().plot()
```

#### Concatenate lines

```python
env['ff_l'] = env['ccs_xl'] + env['ccs_yl']
```

```python
# Inspect
env['ff_l'].survey().plot()
```

#### Mirror lines

```python
env['mirror_ff_l'] = -env['ff_l']
```

```python
env['mirror_ff_l'].survey().plot()
```

#### Assemble a larger machine section

```python
env['quarter'] = (
                      env['mark_ip_mid']
                    + env['experimental_insertion_r']
                    + env['arc_octant']
                    + env['mark_ser_ins_r']
                    + env['service_insertion_r']
                    + env['mark_ser_ins_mid']
                    + env['service_insertion_l']
                    + env['mark_ser_ins_l']
                    + env['arc_octant']
                    + env['experimental_insertion_l'])
```

```python
env['quarter'].survey().plot()
```

#### Assemble the ring

```python
env['fcc_ring'] = 4 * env['fcc_quarter']
```

```python
env['fcc_ring'].survey().plot()
```
